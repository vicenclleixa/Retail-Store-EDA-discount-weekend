### NaN management

```python
df_store_copy.fillna(0, inplace=True)
df_store_copy
```

### Product frequency:

```python
#List of all purchases
all_purchases = df_store_copy.values.flatten()

#Calculate Frequency
product_frequency = pd.Series(all_purchases).value_counts()

#To DataFrame
df_product_frequency = pd.DataFrame({'Product': product_frequency.index, 'Frequency': product_frequency.values})
df_product_frequency = df_product_frequency[df_product_frequency['Product'] != 0]

series_product_frequency = product_frequency.drop(0)
print(series_product_frequency)
```

### DataFrame to wide format

We transform the DataFrame to a wide format which is more conducive to certain types of data analysis and modeling techniques. Rows represent individual customers. Columns represent unique products available for purchase. Cell values are binary indicators: 0.0 indicates the product was not purchased by the customer, and 1.0 indicates the product was purchased.

```python
df_reset = df_store_copy.reset_index().rename(columns={'index': 'CustomerID'})
df_store_melted = df_reset.melt(id_vars='CustomerID', value_name='Product').drop('variable', axis=1)
df_store_melted = df_store_melted.dropna()
df_store_melted['Purchased'] = 1
df_store_melted
```

```python
df_store_melted = df_store_melted.pivot_table(index='CustomerID', columns='Product', values='Purchased', fill_value=0)
df_store_melted
```

### Top 50 most sold products

```python
df_top50_products = df_product_frequency.nlargest(50, 'Frequency')

olors = ['gold'] + ['mediumblue']*(len(df_top50_products)-1)  

plt.figure(figsize=(10, 8))

sns.barplot(x='Frequency', y='Product', data=df_top50_products, palette=colors)

plt.xlabel('Sales Frequency')
plt.ylabel('Product')
plt.title('Top 50 Most Sold Products:', loc='center', fontsize=20)
plt.yticks(fontsize=8)

plt.grid(True, alpha=0.3)

plt.show()
```

### 50 least sold products

```python
df_bottom50_products = df_product_frequency.nsmallest(50, 'Frequency')

sns.set_style("whitegrid")

plt.figure(figsize=(10, 8))
sns.barplot(x='Frequency', y='Product', data=df_bottom50_products, color="dimgrey")

plt.xlabel('Sales Frequency')
plt.ylabel('Product')
plt.yticks(fontsize=8)
plt.title('50 least sold products:', loc='center', fontsize=20)

plt.grid(True, alpha=0.3)

plt.show()
```

### Correlation Map

```python
product_sums = df_store_melted.sum().sort_values(ascending=False)
top_30_products = product_sums.head(30).index
df_top_30 = df_store_melted[top_30_products]

corr_matrix_top_30 = df_top_30.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix_top_30, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, annot_kws={"size": 8}) 
plt.title('Correlation Heatmap of Top 30 Products', fontsize=20)
plt.xticks
plt.yticks
plt.tight_layout() 
plt.show()
```

### Products which are bought together

```python
#Mask to ignore self-correlations
mask = np.triu(np.ones_like(corr_matrix_top_30, dtype=bool))
corr_flat = corr_matrix_top_30.mask(mask)

#Flatten, sort by correlation, and remove NaN values
sorted_corr_pairs = corr_flat.stack().sort_values(ascending=False).dropna()

N = 20
top_pairs = sorted_corr_pairs.head(N)

print("Top most commonly bought pairs of products:")
for pair, corr_value in top_pairs.items():
    print(f"{pair}: Correlation = {corr_value:.2f}")
```

```python
corr_values = top_pairs.values
labels = [f"{x[0]} & {x[1]}" for x in top_pairs.index]

plt.figure(figsize=(15, 8))
plt.bar(labels, corr_values, color='mediumblue')
plt.xlabel('Product Pairs')
plt.ylabel('Correlation')
plt.xticks(rotation=90)
plt.grid(axis='y', alpha=0.75)

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.title('Most Commonly Bought Products', fontsize=20)

plt.show()
```
